```python
import geopy
from geopy.geocoders import Nominatim
n=Nominatim()
```

```python
import pandas
df = pandas.read_csv("Data_Cleaning_Menu_Moodle_data.csv")
df['place']
```

```python
df["Coordinates_place"]=df.ix[0:100, 'place'].apply(n.geocode)
df["Coordinates_location"]=df.ix[0:100, 'location'].apply(n.geocode)
df.head
```

```python
df["place_latitude"] = df.ix[0:100, 'Coordinates_place'].apply(lambda x: x.latitude if x != None else None)
df["place_longitude"] = df.ix[0:100, 'Coordinates_place'].apply(lambda x: x.longitude if x != None else None)
# df.ix[0:10, 'Coordinates_place'].apply(lambda x: x.address if x != None else None).str.extract('(,((.*),))')
df["place_address"] = df.ix[0:100, 'Coordinates_place'].apply(lambda x: x.address if x != None else None)

df["location_latitude"] = df.ix[0:100, 'Coordinates_location'].apply(lambda x: x.latitude if x != None else None)
df["location_longitude"] = df.ix[0:100, 'Coordinates_location'].apply(lambda x: x.longitude if x != None else None)
# df.ix[0:10, 'Coordinates_place'].apply(lambda x: x.address if x != None else None).str.extract('(,((.*),))')
df["location_address"] = df.ix[0:100, 'Coordinates_location'].apply(lambda x: x.address if x != None else None)



```

```python


del df['Coordinates_place']
del df['Coordinates_location']
df
```
